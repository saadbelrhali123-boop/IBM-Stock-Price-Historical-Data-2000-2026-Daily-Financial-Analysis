```python
import pandas as pd
```

```python
df = pd.read_csv('IBM.csv')
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
```

```python
df.sort_index(inplace=True)

print(df.info())
```

```python
import matplotlib.pyplot as plt
```

```python
df['Close'].plot(figsize=(12, 6), title="Évolution du prix de clôture IBM (2000-2026)")
plt.ylabel("Prix ($)")
plt.show()
```

```python
df['Returns'] = df['Close'].pct_change()
```

```python
df['SMA50'] = df['Close'].rolling(window=50).mean()
df['SMA200'] = df['Close'].rolling(window=200).mean()
```

```python
df['SMA50'] = df['Close'].rolling(window=50).mean()
df['SMA200'] = df['Close'].rolling(window=200).mean()
```

```python
df[['Close', 'SMA50', 'SMA200']].plot(figsize=(12, 6), title="IBM - Prix de clôture et Moyennes Mobiles")
plt.legend(["Prix Close", "Moyenne 50j", "Moyenne 200j"])
plt.ylabel("Prix ($)")
plt.show()
```

```python
df['Returns'] = df['Close'].pct_change() * 100
```

```python
df['Returns'].plot(figsize=(12, 4), title="Rendements quotidiens IBM (%)", color='orange')
plt.axhline(0, color='black', lw=1) # Ligne à zéro
plt.ylabel("Variation en %")
plt.show()
```

```python
df['Returns'].hist(bins=100, figsize=(10, 5))
plt.title("Distribution des rendements quotidiens")
plt.show()
```

```python
df['Volume'].plot(figsize=(12, 4), color='green', title="Volume de transactions quotidien")
plt.ylabel("Nombre d'actions")
plt.show()
```

```python
worst_days = df.nsmallest(5, 'Returns')
print("Les 5 plus grosses baisses :")
print(worst_days[['Close', 'Returns']])
```

```python
df['Cumulative_Return'] = (1 + df['Returns'] / 100).cumprod()
```

```python
df['Cumulative_Return'].plot(figsize=(12, 6), title="Croissance d'un investissement de 1$ dans IBM", color='purple')
plt.ylabel("Valeur de l'investissement ($)")
plt.show()
```

```python
df['Volatility'] = df['Returns'].rolling(window=21).std()

df['Volatility'].plot(figsize=(12, 4), title="Volatilité glissante sur 1 mois (21 jours)", color='red')
plt.ylabel("Écart-type des rendements")
plt.show()
```

```python
delta = df['Close'].diff()
gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()

rs = gain / loss
df['RSI'] = 100 - (100 / (1 + rs))
```

```python
plt.figure(figsize=(12, 4))
df['RSI'].plot(title="RSI IBM (Indicateur de momentum)")
plt.axhline(70, color='red', linestyle='--')
plt.axhline(30, color='green', linestyle='--')
plt.ylim(0, 100)
plt.show()
```

```python
import numpy as np
```

```python
df['SMA50'] = df['Close'].rolling(window=50).mean()
df['SMA200'] = df['Close'].rolling(window=200).mean()
```

```python
df['Signal'] = 0.0
df['Signal'] = np.where(df['SMA50'] > df['SMA200'], 1.0, 0.0)
```

```python
df['Strategy_Returns'] = df['Returns'] * df['Signal'].shift(1)
```

```python
df[['Returns', 'Strategy_Returns']].cumsum().apply(np.exp).plot(figsize=(10,6))
plt.title("Performance : Achat Passif vs Stratégie Moyennes Mobiles")
plt.show()
```

```python
rolling_max = df['Close'].cummax()
```

```python
drawdown = (df['Close'] - rolling_max) / rolling_max
```

```python
drawdown.plot(figsize=(12, 4), color='red', kind='area', alpha=0.3)
plt.title("Drawdown Historique IBM (Baisse par rapport au sommet)")
plt.ylabel("Perte en %")
plt.show()

print(f"La pire chute historique a été de : {drawdown.min():.2%}")
```

```python
df['Target'] = df['Close'].shift(-1)
```

```python
data_ml = df[['Close', 'Volume', 'SMA50', 'SMA200', 'RSI', 'Target']].dropna()

print("Aperçu des données prêtes pour le Machine Learning :")
print(data_ml.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
X = data_ml[['Close', 'Volume', 'SMA50', 'SMA200', 'RSI']]
y = data_ml['Target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
mse = mean_squared_error(y_test, predictions)
```

```python
plt.figure(figsize=(12,6))
plt.plot(y_test.index[-100:], y_test.values[-100:], label="Réel")
plt.plot(y_test.index[-100:], predictions[-100:], label="Prédiction", linestyle='--')
plt.title("Prédiction du prix IBM (Modèle Linéaire)")
plt.legend()
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
data_ml['Direction'] = np.where(data_ml['Target'] > data_ml['Close'], 1, 0)
```

```python
X = data_ml[['Close', 'Volume', 'SMA50', 'SMA200', 'RSI']]
y = data_ml['Direction']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
print(f"Précision du modèle (Accuracy) : {accuracy_score(y_test, y_pred):.2%}")
print("\nTableau de performance :")
print(classification_report(y_test, y_pred))
```

```python
best_return = -np.inf
best_params = (0, 0)
```

```python
for fast in range(10, 60, 10):
    for slow in range(100, 250, 20):
        temp_sma_fast = df['Close'].rolling(window=fast).mean()
        temp_sma_slow = df['Close'].rolling(window=slow).mean()
        temp_signal = np.where(temp_sma_fast > temp_sma_slow, 1.0, 0.0)
        temp_returns = (df['Returns']/100) * pd.Series(temp_signal).shift(1).values
        total_return = temp_returns.sum()
        
        if total_return > best_return:
            best_return = total_return
            best_params = (fast, slow)

print(f"Meilleure combinaison : SMA {best_params[0]} / SMA {best_params[1]}")
print(f"Rendement total optimisé : {best_return:.2%}")
```

```python
n_simulations = 1000
n_days = 252
last_price = df['Close'].iloc[-1]
```

```python
results = np.zeros((n_days, n_simulations))

for i in range(n_simulations):
    daily_returns = np.random.normal(df['Returns'].mean()/100, df['Returns'].std()/100, n_days)
    price_series = last_price * (1 + daily_returns).cumprod()
    results[:, i] = price_series
```

```python
plt.figure(figsize=(10,6))
plt.plot(results, color='gray', alpha=0.1)
plt.plot(results.mean(axis=1), color='red', linewidth=2, label='Moyenne des scénarios')
plt.title(f"Simulation de Monte Carlo : Futurs possibles d'IBM sur 1 an")
plt.xlabel("Jours")
plt.ylabel("Prix ($)")
plt.legend()
plt.show()
```
